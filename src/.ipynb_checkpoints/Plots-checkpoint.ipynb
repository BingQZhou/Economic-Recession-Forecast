```python
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt
from datetime import datetime
from utils import *
```

```python
# import data
consumer_staples = pd.read_csv("../data/S&P 500 Consumer Staples Historical Data.csv")
sp500 = pd.read_csv("../data/S&P 500 Historical Data.csv")
yield_rate = pd.read_excel("../data/yield.xlam")
```

## Data Cleaning

```python
consumer_staples['Date'] = consumer_staples['Date'].apply(investing_convert_time)
sp500['Date'] = sp500['Date'].apply(investing_convert_time)
sp500['Price'] = [float(str.replace(',', '')) for str in sp500['Price']]
yield_rate['Date'] = yield_rate['Date'].apply(yield_convert_time)
display(consumer_staples.head())
display(sp500.head())
display(yield_rate)
```

```python
# consumer and sp500
combined = consumer_staples[['Date', 'Price']].merge(sp500[['Date', 'Price']], on = "Date", how = "inner", suffixes=('_consumer', '_sp500'))
combined['Price_consumer'] = normalize(combined['Price_consumer'])
combined['Price_sp500'] = normalize(combined['Price_sp500'])
display(combined.head())
```

```python
# yield rate and sp500
yield_combined = yield_rate[['Date', '2 Yr', '10 Yr']].merge(sp500[['Date', 'Price']], on = "Date", how = "inner", 
                                                             suffixes=('', '_sp500'))
yield_combined.iloc[:, 1:4] = yield_combined.iloc[:, 1:4].apply(normalize)
yield_combined.head()
```

```python
# consumer staples plot
fig = plt.figure(figsize = (20, 8))
plt.xticks(np.arange(0, len(combined), step = 50))
seaborn.lineplot(data = combined, x = "Date", y = "Price_consumer")
seaborn.lineplot(data = combined, x = "Date", y = "Price_sp500")
plt.show()
```

```python
# yield rate plot
fig = plt.figure(figsize = (20, 8))
plt.xticks(np.arange(0, len(yield_combined), step = 200))
seaborn.lineplot(data = yield_combined, x = "Date", y = "2 Yr")
seaborn.lineplot(data = yield_combined, x = "Date", y = "10 Yr")
seaborn.lineplot(data = yield_combined, x = "Date", y = "Price")
plt.legend(['2 year', '10 year', 'sp500'])
plt.show()
```
